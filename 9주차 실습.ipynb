```python
#1번
def square(n):
    k = n**2
    return k
###테스트
print(square(5))


#2번
def max2(m,n):
    if m > n:
        return m
    else:
        return n
def min2(m,n):
    if m < n:
        return m
    else:
        return n
###테스트
max2(10,20)
min2(10,20)


#3번
def mile2km(m):
    k = m*1.61
    return k
###테스트
for i in range(5):
    print(mile2km(i+1))


#4번
def is_prime(n):
    if n == 1:
            return False
    i = 2
    while n > i:
        if n % i == 0:
            return False
        i+=1
    return True
'''
테스트
for i in range(1000000):
    if is_prime(i+1) == True:
        print(i+1)
'''


#5번
def statistical_numbers(*numbers):
    mytuple=()
    hop=0
    mylist=[]
    if numbers == 0:
        print('종료')
        return
    else:
        mylist.extend(numbers)
    for i in mylist:
        hop+=i
    평균 = hop/len(mylist)   #평균구하기
    big=mylist[0]     #최댓값 구하기
    small=mylist[0]    #최솟값 구하기
    for i in mylist:
        if i>big:
            big=int(i)
    for i in mylist:
        if i<small:
            small=int(i)

    mytuple=(평균,big,small)
    return mytuple
###테스트
result=statistical_numbers(10,20,30,40,50,3,90,33,102,3201)
print(result)

#6번
def is_palindrome(str_):
    str_=str_.lower()
    for i in range(len(str_)//2):
        if str_[i]==str_[-i-1]:
            continue
        else:
            print('회문이 아닙니다')
            return
    print('회문입니다')
###테스트
is_palindrome('Civic')
is_palindrome('Anna')
is_palindrome('Kayak')
is_palindrome('Level')
is_palindrome('My gym')

#7번
def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year %400 == 0:
                print('윤년입니다')
            else:
                print('평년입니다')
        print('윤년입니다.')
    else:
        print('평년입니다.')

#8번
def recursive_sum(n, m):
    if n == m:
        return n
    else:
        return n + recursive_sum(n+1, m)

###테스트
print(recursive_sum(4,10))

#9번
def hanoy(layer,start, helper, end):
    if layer == 1:
        print(f'1원반을 {start}에서 {end}로 옮겨라.')
        return
    hanoy(layer -1 ,start, end, helper) 
    print(f'{layer}원반을 {start}에서 {end}로 옮겨라. ')
    hanoy(layer -1,helper, start, end)
###테스트
n= int(input('원판 갯수 입력:'))
hanoy(n,'A','B','C')

#10번
def check_password(str_):
    if len(str_) < 8:
        return False
    has_upper = False
    has_lower = False
    has_digit = False
    for i in str_:
        if i.isupper():
            has_upper = True
        elif i.islower():
            has_lower = True
        elif i.isdigit():
            has_digit = True
    if has_upper and has_lower and has_digit:         #각 변수들은 True False로 설정되어 있으면 따로 다른 조건필요 없이 그냥 if만 쓰면 됨!
        return True
    else:
        return False
###테스트
print(check_password('Hello111'))
print(check_password('hello111'))
print(check_password('Helloooo'))
print(check_password('HELLO111'))

#11번
def bestdivide(n,m):
    divide = []
    if n < m:
        for i in range(n):
            if n % (i+1) == 0:
                if m % (i+1) == 0:
                    divide.append(i+1)
    else:
        for i in range(m):
            if m % (i+1) == 0:
                if n % (i+1) == 0:
                    divide.append(i+1)
    divide.sort()
    print(divide[-1])  

###테스트
bestdivide(40,120)

#12번
def dec2bin(n):
    if n == 0:
        return '0'
    
    # 결과를 담을 변수
    binary_str = ''
    

    # n이 0이 될 때까지 반복하여 이진수 계산
    while n > 0:
        binary_str = str(n % 2) + binary_str
        n = n // 2
    return binary_str

###테스트
print(dec2bin(10))
```
